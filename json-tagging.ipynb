```python
r = requests.get('https://scripts.njae.me.uk/festive-road-pictures/small/2017/festival-of-giants-2017/festival-of-giants.json')
r.json()
```

```python
with open('festival-of-giants-hand-tags.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    csv_rows = [row for row in reader]
len(csv_rows)
```

```python
csv_rows[1]
```

```python
compressed = []
for r in csv_rows:
    if r['Imaage URL'] != '':
        cr = {'url': 'https://scripts.njae.me.uk/festive-road-pictures/small/2017/festival-of-giants-2017/' + r['Imaage URL'], 'keys': []}
        for k, v in r.items():
            if v == 'x':
                cr['keys'] += [k.lower()]
    compressed += [cr]
compressed
```

```python
with open('festival-of-giants-2017-hand-tags.json', 'w') as outfile:
    json.dump(compressed, outfile, indent=2)
```
