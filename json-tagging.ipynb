```python
with open('festival-of-giants-2017-hand-tags.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    csv_rows = [row for row in reader]
len(csv_rows)
```

```python
csv_rows[1]
```

```python
compressed = []
for r in csv_rows:
    if r['Imaage URL'] != '':
        cr = {'url': 'https://scripts.njae.me.uk/festive-road-pictures/small/2017/festival-of-giants-2017/' + r['Imaage URL'], 'hand-tags': []}
        for k, v in r.items():
            if v == 'x':
                cr['hand-tags'] += [k.lower()]
    compressed += [cr]
compressed
```

```python
with open('festival-of-giants-2017-hand-tags.json', 'w') as outfile:
    json.dump(compressed, outfile, indent=2)
```

```python
with open('festival-of-giants-2017-hand-tags-with-generated.json') as infile:
    machine_tagged = json.load(infile)
machine_tagged[0]
```

```python
machine_tags_for_hand = collections.defaultdict(collections.Counter)
for mts in machine_tagged:
    for ht in mts['keys']:
        machine_tags_for_hand[ht].update(t for t in mts['tags'])
machine_tags_for_hand
```

```python
for t in machine_tags_for_hand:
    print(t, machine_tags_for_hand[t].most_common(5))
```
